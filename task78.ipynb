```
def read_file_to_string(file_path):
    try:
        # Open the file in read mode and read the content
        with open(file_path, 'r', encoding='utf-8') as file:
            file_content = file.read()
        return file_content
    except FileNotFoundError:
        print(f"The file at {file_path} was not found.")
        return None
    except IOError:
        print(f"An error occurred while reading the file at {file_path}.")
        return None

# Example usage
file_path = '/content/sample_data/MERGED.txt'  # Replace with your file path
text = read_file_to_string(file_path)

if text:
    print(text)
```

```
import nltk
import string
from nltk.corpus import stopwords
from nltk.util import ngrams
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Function to preprocess text: Remove punctuation and stopwords
def preprocess_text(text):
    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])

    # Tokenize the text into words
    words = nltk.word_tokenize(text.lower())  # Convert to lowercase to ensure case-insensitivity

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    return words

# Function to count word frequency
def count_words(text):
    words = preprocess_text(text)
    word_freq = Counter(words)
    return word_freq

# Function to count n-grams frequency
def count_ngrams(text, n=2):
    words = preprocess_text(text)
    n_grams = ngrams(words, n)
    ngram_freq = Counter(n_grams)
    return ngram_freq

# Count word frequency
word_count = count_words(text)
print("Word Frequency:")
print(word_count)

# Count bigram frequency (2-grams)
bigram_count = count_ngrams(text, n=2)
print("\nBigram Frequency (2-grams):")
print(bigram_count)

# Count trigram frequency (3-grams)
trigram_count = count_ngrams(text, n=3)
print("\nTrigram Frequency (3-grams):")
print(trigram_count)

# Alternatively, using CountVectorizer from sklearn to count word frequencies
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform([text])
word_freq_sklearn = dict(zip(vectorizer.get_feature_names_out(), X.toarray().flatten()))
print("\nWord Frequency using sklearn CountVectorizer:")
print(word_freq_sklearn)

```

```
import re

words = preprocess_text(text)

def pos_tagging(words):
    pos_tags = []

    for word in words:
        if re.match(r'^\w+ed$', word):  # simple pattern for past tense verbs
            pos_tags.append((word, 'VERB'))
        elif re.match(r'^\w+ing$', word):  # pattern for present tense verbs
            pos_tags.append((word, 'VERB'))
        elif re.match(r'^\w+s$', word):  # pattern for plural nouns
            pos_tags.append((word, 'NOUN'))
        elif re.match(r'^\w+y$', word):  # pattern for adjectives ending in 'y'
            pos_tags.append((word, 'ADJ'))
        else:
            pos_tags.append((word, 'unknown'))  # Default to unknown

    return pos_tags

tagged_words = pos_tagging(words)

# Display the results
print(tagged_words)
```

```
# Code snippet #8.7 Sentiment analysis with NLTK
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon
nltk.download('vader_lexicon')

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Sample text
texts = [text]

# Analyze sentiment
for text in texts:
    sentiment_score = sia.polarity_scores(text)
    sentiment = "Neutral"

    if sentiment_score['compound'] >= 0.05:
        sentiment = "Positive"
    elif sentiment_score['compound'] <= -0.05:
        sentiment = "Negative"

    print(f"Text: {text}\nSentiment: {sentiment}\nScores: {sentiment_score}\n")
```
